```python
import pandas
import torch
from catboost import CatBoostClassifier
from pandas import notnull, DataFrame
```

```python
sources = pandas.read_csv('../input/newsclass01/train.csv', index_col='Id')[['source']]
data = pandas.read_csv('../input/news-text-model-output/text_model_train.csv', index_col='Id').join(pandas.read_csv('../input/news-image-model-output-1617229768/image_model_train_1617229768.csv', index_col='Id'))
```

```python
data.head()
```

```python
sources.head()
```

```python
classifier = CatBoostClassifier(metric_period=100, iterations=5000, nan_mode='Max')
```

```python
classifier.fit(data, sources)
```

```python
test_data = pandas.read_csv('../input/news-text-model-test-output/text_model_test.csv', index_col='Id').join(pandas.read_csv('../input/news-image-model-test-output-1617230717/image_model_test_1617230717.csv', index_col='Id'))
```

```python
test_data.head()
```

```python
data_frame = DataFrame(classifier.predict(test_data), index=test_data.index, columns=['Predicted']).rename_axis('Id')
data_frame.to_csv(f'./submission.csv')
```
